```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
import pandas as pd

df_complaints=pd.read_csv('../drive/MyDrive/Colab Notebooks/Complaint/complaints.csv',low_memory=False)
desc_col='Consumer complaint narrative'

df_complaints['Year']=pd.to_datetime(df_complaints['Date received']) .dt.year
df_complaints['Month']=pd.to_datetime(df_complaints['Date received']) .dt.month
df_complaints['Day']=pd.to_datetime(df_complaints['Date received']) .dt.day

text_splitter = RecursiveCharacterTextSplitter(
    # Set a really small chunk size, just to show.
    chunk_size = 1000,
    chunk_overlap  = 40,
    length_function = len,
)

def summarize_doc(text_splitter,llm_chain,doc):
    try:
        texts = text_splitter.create_documents([doc])
        summary = llm_chain.run(texts)
    except:
         print(doc)
         print('*****************')
         summary="@@@Error@@@ "+str(doc)
   # print(summary)
    return summary

# df_complaints['complaint_summary']=df_complaints[desc_col].apply(lambda doc: summarize_doc (text_splitter,chain,doc) )


c=[]
cnt=0
for idx,cmplnt in df_complaints.iterrows():
      c.append(summarize_doc (text_splitter,chain,cmplnt[desc_col]))
      print(cnt)
      cnt=cnt+1

df_complaints['complaint_summary']=c

df_complaints.to_csv('../drive/MyDrive/Colab Notebooks/Complaint/langchain_complaints.csv')
```
